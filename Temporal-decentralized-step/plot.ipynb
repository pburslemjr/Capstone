```python
from matplotlib import pyplot as plt
import re

f = open("rewards_2.txt", "r")
f_2 = open("model_used_1.txt", "r")
x = []
y = []
y_2 = []

data = f.readlines()
data_2 = f_2.readlines()
rd = []
total_rew = 0.0
for i, line in enumerate(data):
    if(i<0):
        continue
    rd.append(re.findall("-?\d+\.*\d*", line))
    if(rd[-1] == []):
        break
    if(int(data_2[i]) == 1):#-2.5 < float(rd[-1][0]) < 2.5):
        x.append(i)
        y.append(float(rd[-1][0]))
    total_rew += y[-1]
    if(i%50 == 0):
        print(total_rew)
        y_2.append(total_rew)
        total_rew  = 0.0
print(total_rew, len(y)%50)
y_2.append(total_rew)
fig = plt.figure(figsize=(8,8))
plt.subplot(2,1,1)
plt.plot(x,y)
plt.subplot(2,1,2)
plt.plot(range(len(y_2)),y_2)

```
