```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
%matplotlib inline

data = pd.read_csv('../TC_E1/rewards.txt',sep='\s+',header=None)
data = pd.DataFrame(data)


x = data[1]
y = data[2]
#x = [x for x in range(len(y))]
frac = data[3]

average = []
average_indices = []

for i in range(math.floor(float(len(y))/50.0) + 1):
    if i > 0:
        average.append(np.mean(y[(i-1)*50:i*50]))
        average_indices.append(i*50)
plt.figure(figsize=(15, 10))

for i,f in enumerate(frac):
    if i > 0 and f > frac[i-1]:
        plt.axvline(x=i, color='lightgreen')

plt.plot(x, y,'lightblue')
plt.plot(average_indices, average,'r')

plt.axhline(y=0.0, color='salmon', linestyle='--')
plt.show()
plt.figure(figsize=(15, 5))
plt.plot(x, frac,'r')
plt.show()
```
